## ƒåas≈• 1: LLM vs Agent

### ƒåo je LLM (Language Model)?
- Predpoved√° ƒèal≈°ie slovo
- Jednorazov√° odpoveƒè
- ≈Ωiadne akcie

### ƒåo je Agent?
- LLM + schopnos≈• kona≈•
- Vie pou≈æ√≠va≈• n√°stroje
- M√¥≈æe pl√°nova≈• kroky
- Iteruje a≈æ k√Ωm nedosiahne cieƒæ

````
#@title üìù Pr√≠klad: ƒåist√© LLM

import google.generativeai as genai

genai.configure(api_key=os.environ["GOOGLE_API_KEY"])
model = genai.GenerativeModel('gemini-2.0-flash')

# LLM len odpoved√° - nevie "kona≈•"
response = model.generate_content("Ak√© je aktu√°lne poƒçasie v Prahe?")
print("üî§ LLM odpoveƒè:")
print(response.text)
print("\n‚ö†Ô∏è V≈°imni si: LLM nevie skutoƒçne zisti≈• poƒçasie - len h√°d≈æe.")
````

## ƒåas≈• 2: Prv√Ω Agent

Agent m√¥≈æe:
1. **Rozm√Ω≈°ƒæa≈•** - analyzova≈• √∫lohu
2. **Pl√°nova≈•** - rozdeli≈• na kroky
3. **Kona≈•** - pou≈æi≈• n√°stroje
4. **Pozorova≈•** - vyhodnoti≈• v√Ωsledok
5. **Opakova≈•** - a≈æ k√Ωm nie je hotovo

````
#@title ü§ñ Vytvorenie z√°kladn√©ho agenta

from smolagents import CodeAgent
from smolagents.models import LiteLLMModel

# Inicializ√°cia modelu
model = LiteLLMModel(
    model_id="gemini/gemini-2.0-flash",
    api_key=os.environ["GOOGLE_API_KEY"]
)

# Vytvorenie agenta
agent = CodeAgent(
    tools=[],           # Zatiaƒæ bez n√°strojov
    model=model,
    max_steps=3         # Max poƒçet krokov
)

print("‚úÖ Agent vytvoren√Ω!")
print(f"   Model: gemini-2.0-flash")
print(f"   Max krokov: 3")
print(f"   N√°stroje: ≈æiadne")
````

````
#@title ‚ñ∂Ô∏è Spustenie agenta

# Jednoduch√° √∫loha
task = "Vypoƒç√≠taj koƒæko dn√≠ m√° rok 2026 a ƒçi je priestupn√Ω."

print(f"üìã √öloha: {task}")
print("="*50)

result = agent.run(task)

print("="*50)
print(f"‚úÖ V√Ωsledok: {result}")
````

## ƒåas≈• 3: Pozorovanie agenta

Agent pracuje v cykloch:
```
Thought ‚Üí Action ‚Üí Observation ‚Üí Thought ‚Üí ...
```

Toto sa vol√° **ReAct pattern** (Reasoning + Acting).

````
#@title üîç Agent s verbose v√Ωstupom

# Vytvor√≠me nov√©ho agenta s verbose=True
verbose_agent = CodeAgent(
    tools=[],
    model=model,
    max_steps=5,
    verbosity_level=2  # Uk√°≈æe detailn√© kroky
)

print("üîç Sleduj ako agent rozm√Ω≈°ƒæa:")
print("="*50)

result = verbose_agent.run("Vytvor zoznam 5 eur√≥pskych hlavn√Ωch miest a ich kraj√≠n.")

print("="*50)
print(f"\n‚úÖ Fin√°lny v√Ωsledok: {result}")
````

## ƒåas≈• 4: System Prompt

System prompt definuje "osobnos≈•" a pravidl√° agenta.

````
#@title üé≠ Agent s vlastn√Ωm system promptom

# Custom system prompt
custom_prompt = """
Si priateƒæsk√Ω AI asistent pre Kyndryl AI Academy.

Pravidl√°:
1. Odpovedaj struƒçne a jasne
2. Pou≈æ√≠vaj pr√≠klady
3. Ak nieƒço nevie≈°, povedz to priamo
4. Na konci v≈ædy polo≈æ jednu dopl≈àuj√∫cu ot√°zku
"""

custom_agent = CodeAgent(
    tools=[],
    model=model,
    max_steps=3,
    system_prompt=custom_prompt
)

result = custom_agent.run("ƒåo je to prompt engineering?")
print(result)
````
